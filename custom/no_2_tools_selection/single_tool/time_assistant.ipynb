1) **Define tools**  
2) **Provide tools to LLMs**
3) **Execute code from LLMs**

### Step 1: Define tools

```python
from datetime import datetime
from zoneinfo import ZoneInfo

def get_current_time(timezone: str = "UTC") -> str:
    """
    Returns the current time in the specified timezone as a string formatted as HH:MM:SS.
    """
    timezone_obj = ZoneInfo(timezone)
    current_time = datetime.now(tz=timezone_obj)
    return current_time.strftime("%H:%M:%S")
```

### Step 2: Provide tools to LLMs

```python
def weather_assistant_prompt(region: str = "UTC") -> str:
    prompt = f"""
    You have access to a tool called get_current_time that returns the current time for a specified timezone.
    To use this tool, call it with the desired timezone as an argument as follows: 

    FUNCTION CALL:
    get_current_time("Timezone").

    User Message: What is the current time in {region}?
    """
    return prompt
```

```python
model_name = "gemini-2.5-flash"
response = utils.get_response(model_name, weather_assistant_prompt("Hanoi, Vietnam"))
print("Model Response:", response, sep="\n")
```

### Step 3: Execute code

```python

if "FUNCTION CALL:" in response:
    print("Code executed:")
    function_call_match = re.search(r'FUNCTION CALL:\s*(\w+)\("([^"]+)"\)', response)
    if function_call_match:
        function_name = function_call_match.group(1)
        argument = function_call_match.group(2)

        if function_name == "get_current_time":
            current_time = get_current_time(argument)
            final_response = f"The current time in {argument} is {current_time}."
            print(final_response)
```
